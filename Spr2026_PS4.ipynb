```python
# Libraries
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
# Get data from dropbox
# Import data from dropbox
!wget -O food_health.csv https://www.dropbox.com/scl/fi/a3jpecmxq1i9tkdesjwtp/food_health.csv?rlkey=ylj48ghu4lvxi89lvupn8k2xn&dl=0

# Import
data = pd.read_csv("food_health.csv")
print(data.head())
```

```python
# Summary statistics for two baseline variables
print(data[['bp_systolic', 'sport_days']].describe())

# Need to clean out invalid values (sport_days = 99)
data = data[data['sport_days'] < 99]

# There are also missing values in bp_systolic
# The count is less than the total sample size
data = data[~data['bp_systolic'].isna()]

# Let's check
print('\n')
print(data[['bp_systolic', 'sport_days']].describe())
```

```python
# Simple regression
X = sm.add_constant(data['sport_days'])
y = data['bp_systolic']

model = sm.OLS(y, X).fit(cov_type='HC3')
print(model.summary())
```

# Case 2: Scaling A/B Tests Under Uncertainty

Imagine you are part of the experimentation team at an e-commerce firm that runs a large number of A/B tests each year. Each experiment compares a treatment (a product, pricing, messaging, or user-experience change) to a control and reports summary results from the test. The firm cannot treat every positive-looking result as a guaranteed win, and the stakes vary across experiments: some changes are small and low-risk, while others affect large parts of the customer experience and could have much larger upside (or downside).

In this case, you will work with a synthetic dataset of several thousand experiments. For each experiment, you are given standard A/B test summary statistics, including sample average profit in the control and treatment arms, sample standard deviations, sample sizes, standard errors, the standard error of the difference in means, and a p-value for testing whether the average treatment effect is zero. You will also see additional descriptors of the experiment (for example, the type of platform feature being tested) and a projected future deployment scale, which captures how many customers would be affected if the treatment were rolled out.

Specifically, the variables in the data are

* `experiment_id`: Unique identifier for the experiment.
* `dataset_split`: Indicates which dataset split the row belongs to. You can safely ignore this variable.
* `experiment_area`: Broad area of the business/platform affected by the experiment.
* `intervention_type`: Type of intervention being tested (for example, small UI tweak, major UX change, pricing rule change, recommendation change, etc.).
* `target_segment`: Customer segment targeted by the experiment (for example, new users, returning users, high-value users, etc.).
* `ux_disruption_score`: A numeric score capturing how disruptive the treatment is to the user experience (higher values indicate a larger or more noticeable change).
* `implementation_cost_index`: A numeric index for the cost/complexity of implementing the treatment at scale (higher values indicate more costly implementation). **Important: All outcomes are already net of costs.**
* `team_owner`: Team responsible for the experiment (for example, Growth, Checkout, Search, CRM, etc.).
* `experiment_month`: Month in which the experiment was run.
* `analyst_tenure_months`: Tenure of the analyst running the experiment, measured in months.
* `control_mean_profit`: Sample average profit per customer in the control group. **Important: This IS profit. Costs have been incorporated.**
* `treatment_mean_profit`: Sample average profit per customer in the treatment group.  **Important: This IS profit. Costs have been incorporated.**
* `control_sd_profit`: Sample standard deviation of profit per customer in the control group.
* `treatment_sd_profit`: Sample standard deviation of profit per customer in the treatment group.
* `n_control`: Number of observations (customers) in the control group.
* `n_treatment`: Number of observations (customers) in the treatment group.
* `se_control_mean`: Standard error of the control-group sample mean.
* `se_treatment_mean`: Standard error of the treatment-group sample mean.
* `diff_in_means`: Estimated average treatment effect (ATE), computed as treatment_mean_profit - control_mean_profit.
* `se_diff_in_means`: Standard error of diff_in_means (used for inference on the ATE).
* `p_value_ate_zero`: p-value for testing the null hypothesis that the average treatment effect is zero.
* `future_deploy_scale`: Projected number of customers affected if the treatment is rolled out (used to think about the impact of scaling the treatment).

Your ultimate goal is not just to identify “statistically significant” experiments. Instead, the goal is to think carefully about what a firm should consider when deciding which experiments to scale. In other words: how should a decision rule use estimated effects, uncertainty, scale, and other experiment features when making implementation decisions?

* `hh_id`: household/customer identifier.

* `spend`: realized spending.
* `income`, `age`, `family_size`, `sex`, `race`, `prev_interactions`: pre-treatment household features.

We can define household-level profit as

`profit` = (`price` - `cost`)*`purchase`.

### Sanity checks

```python
# Randomization ratio
treat_share = df[COL_TREAT].mean()
print(f"Share treated (A=1): {treat_share:.3f}")

# Covariate balance: means by group
num_covs = [c for c in COVARIATES if df[c].dtype != "object"]

balance_num = df.groupby(COL_TREAT)[num_covs].mean()
balance_num.index = ["Control (A=0)", "Treatment (A=1)"]

print("Mean covariates by group")
display(balance_num)

# Two-sample t-tests for balance
balance_tests = []

for col in num_covs:

    control = df[df[COL_TREAT] == 0][col]
    treat = df[df[COL_TREAT] == 1][col]

    t_stat, p_val = stats.ttest_ind(treat, control, equal_var=False)

    balance_tests.append({
        "variable": col,
        "control_mean": control.mean(),
        "treat_mean": treat.mean(),
        "difference": treat.mean() - control.mean(),
        "t_stat": t_stat,
        "p_value": p_val
    })

balance_tests = pd.DataFrame(balance_tests).sort_values("p_value")

print("T-tests for equality of means (treatment vs control)")
display(balance_tests)
```

### A Baseline Heterogeneity Model

We now move to a simple targeting exercise.

A reasonable starting point is to estimate the discount's effect on profit for different types of households. To keep things simple, we will use one approach only:

1. Split the data into a training sample and a holdout sample.
2. Fit one model for expected profit under discount and one model for expected profit under regular price.
3. For each household in the holdout sample, estimate the household-level profit lift from the discount.
4. Use these estimates to build a simple targeting rule.


```python
# Train/holdout split
train_df, holdout_df = train_test_split(
    df,
    test_size=0.30,
    random_state=42,
    stratify=df[COL_TREAT]
)

# One-hot encode categorical covariates
X_train = pd.get_dummies(train_df[COVARIATES], columns=[c for c in ["sex", "race"] if c in COVARIATES], drop_first=True)
X_holdout = pd.get_dummies(holdout_df[COVARIATES], columns=[c for c in ["sex", "race"] if c in COVARIATES], drop_first=True)
X_holdout = X_holdout.reindex(columns=X_train.columns, fill_value=0)

# Separate treated and control observations in the training sample
train_t = train_df[train_df[COL_TREAT] == 1]
train_c = train_df[train_df[COL_TREAT] == 0]

Xt = pd.get_dummies(train_t[COVARIATES], columns=[c for c in ["sex", "race"] if c in COVARIATES], drop_first=True).reindex(columns=X_train.columns, fill_value=0)
Xc = pd.get_dummies(train_c[COVARIATES], columns=[c for c in ["sex", "race"] if c in COVARIATES], drop_first=True).reindex(columns=X_train.columns, fill_value=0)

Yt = train_t["profit"].values
Yc = train_c["profit"].values

# Fit separate models for profit under each treatment state
m1 = RandomForestRegressor(n_estimators=400, min_samples_leaf=40, random_state=42)
m0 = RandomForestRegressor(n_estimators=400, min_samples_leaf=40, random_state=42)

m1.fit(Xt.values, Yt)
m0.fit(Xc.values, Yc)

# Estimated profit lift from offering the discount
mu1_hat = m1.predict(X_holdout.values)
mu0_hat = m0.predict(X_holdout.values)
tau_hat = mu1_hat - mu0_hat

holdout = holdout_df.copy()
holdout["tau_hat_profit"] = tau_hat

print("Mean estimated incremental profit from discount:", round(holdout["tau_hat_profit"].mean(), 4))
print("Percentiles of estimated lift:")
print(holdout["tau_hat_profit"].quantile([0.1, 0.25, 0.5, 0.75, 0.9]))
```

```python
# Inspect the distribution of predicted profit lift
plt.figure(figsize=(7, 4))
plt.hist(holdout["tau_hat_profit"], bins=40)
plt.title("Estimated profit lift from discount on holdout sample")
plt.xlabel("Estimated treatment effect on profit")
plt.ylabel("Count")
plt.grid(True)
plt.show()

# Simple segment summary: top vs bottom predicted-lift quartiles
holdout["tau_quartile"] = pd.qcut(holdout["tau_hat_profit"], q=4, labels=["Q1 (lowest)", "Q2", "Q3", "Q4 (highest)"])
segment_summary = holdout.groupby("tau_quartile")[["tau_hat_profit", "income", "age", "family_size", "prev_interactions"]].mean()
display(segment_summary)
```

### A Simple Targeting Rule

A natural first targeting rule is:

> Offer the discount only to households whose predicted profit lift is positive

The next code cell evaluates this rule on the holdout sample. Because treatment was randomized, we can estimate the value of any targeting rule using *inverse-probability weighting*. For comparison, we also estimate the value of two benchmark policies:

1. Never discount
2. Discount everyone
3. Discount only if predicted lift is positive

```python
# Randomization probability in the experiment
p_treat = df[COL_TREAT].mean()

targeting_cost = 0.0  # Allow for an additional targeting cost

# Candidate policies on the holdout sample
holdout["pi_none"] = 0
holdout["pi_all"] = 1
holdout["pi_target"] = (holdout["tau_hat_profit"] > targeting_cost).astype(int)

def ipw_policy_value(
    data,
    policy_col,
    outcome_col="profit",
    treat_col=COL_TREAT,
    p=p_treat,
    targeting_cost=targeting_cost
):
    chosen = data[policy_col]
    A = data[treat_col]
    Y = data[outcome_col]

    # Net outcome under the policy = profit minus targeting cost if targeted
    Y_net = Y - targeting_cost * chosen

    weights = np.where(A == 1, 1 / p, 1 / (1 - p))
    return np.mean((chosen == A) * weights * Y_net)

policy_values = pd.DataFrame({
    "policy": ["Never discount", "Discount everyone", "Target if predicted lift > cost"],
    "estimated_profit_per_household": [
        ipw_policy_value(holdout, "pi_none"),
        ipw_policy_value(holdout, "pi_all"),
        ipw_policy_value(holdout, "pi_target")
    ],
    "discount_rate": [
        holdout["pi_none"].mean(),
        holdout["pi_all"].mean(),
        holdout["pi_target"].mean()
    ]
})

display(policy_values.sort_values("estimated_profit_per_household",
                                  ascending=False))
```
